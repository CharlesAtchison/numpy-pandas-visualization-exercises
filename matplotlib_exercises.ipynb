# Mathplotlib Exercises

### 1. Use matplotlib to plot the following equation:
* $
\begin{align}
y=x^2 - x + 2
\end{align}
$

```python
import matplotlib.pyplot as plt
x = list(range(1,100))
y = [(n ** 2 - n + 2) for n in x]
plt.figure(figsize=(14, 6))
plt.scatter(x, y)
plt.plot(x,y, ls=':')
plt.title('$y = x^2 - x + 2$')
plt.annotate('Origin', xy=(0,0), xytext=(20, 5000), arrowprops={'facecolor': 'red'})
plt.show()
```

### $
\begin{align}
y=\sqrt{x}
\end{align}
$

```python
x = list(range(1,100))
y = [(n ** .5) for n in x]
plt.figure(figsize=(14, 6))
plt.scatter(x, y)
plt.plot(x,y, ls=':')
plt.title('$y = \u221Ax$')
plt.show()
```

### $

$

```python
x = list(range(1,100))
y = [(n ** 3) for n in x]
plt.figure(figsize=(14, 6))
plt.scatter(x, y)
plt.plot(x,y, ls=':')
plt.title('$y = x^3$')
plt.show()
```

### $
\begin{align}
y=2^x
\end{align}
$

```python
x = list(range(1,10))
y = [(2 ** n) for n in x]
plt.figure(figsize=(14, 6))
plt.scatter(x, y)
plt.plot(x,y, ls=':')
plt.title('$y = 2^x$')
plt.show()
```

### $
\begin{align}
y=\frac{1}{(x+1)}
\end{align}
$

```python
x = list(range(1,30))
y = [(1/(n+1)) for n in x]
plt.figure(figsize=(14, 6))
plt.scatter(x, y)
plt.plot(x,y, ls=':')
plt.title('$y = \dfrac{1}{(x+1)}$')
plt.show()
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots

```python
plt.figure(figsize=(14, 6))
x = list(range(1,10))

y1 = [(n ** .5) for n in x]
y2 = [(n ** 3) for n in x]
y3 = [(2 ** n) for n in x]
y4 = [(1/ (n + 1)) for n in x]
y_list = [y1, y2, y3, y4]
titles = ['$y = \u221Ax$', '$y = x^3$', '$y = 2^x$', '$y = \dfrac{1}{(x+1)}$']

for i, y_val in enumerate(y_list):
    plt.subplot(1, 4, i+1)
    plt.plot(x, y_val)
    plt.title(titles[i])

    
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize=(14, 6))
x = list(range(0,11))

y1 = [(n ** .5) for n in x]
y2 = [(n ** 3) for n in x]
y3 = [(2 ** n) for n in x]
y4 = [(1/ (n + 1)) for n in x]
y_list = [y1, y2, y3, y4]
titles = ['$y = \u221Ax$', '$y = x^3$', '$y = 2^x$', '$y = \dfrac{1}{(x+1)}$']

for i, y_val in enumerate(y_list):
    plt.plot(x, y_val, label=titles[i])

plt.ylim(0,100)
plt.xlim(0,10)
plt.suptitle('4 Equations combined')
plt.legend()
plt.show()
```
