```python
import matplotlib.pyplot as plt
import math
import pandas as pd
import numpy as np
```

$
\begin{align}
y = 0n + 1
\end{align}
$

```python
def first_eq(n):
    return (0.0 * n) + 1.0
```

$
\begin{align}
y = log(n)
\end{align}
$

```python
def second_eq(n):
    try:
        return math.log(float(n))
    except:
        return None
```

$
\begin{align}
y = n
\end{align}
$

```python
def third_eq(n):
    return float(n)
```

$
\begin{align}
y = n * log(n)
\end{align}
$

```python
def fourth_eq(n):
    try:
        return float(n) * math.log(float(n))
    except:
        return None
```

$
\begin{align}
y = n^2
\end{align}
$

```python
def fifth_eq(n: float):
    return float(n) ** 2.0
```

$
\begin{align}
y = 2^n
\end{align}
$

```python
def sixth_eq(n):
    return 2.0 ** float(n)
```

$
\begin{align}
y = n!
\end{align}
$

```python
def seventh_eq(n):
    return math.gamma(n)
```

$
\begin{align}
y = n^n
\end{align}
$

```python
def eighth_eq(n):
    return float(n) ** float(n)
```

```python
plt.figure(figsize=(14, 6))

big_o_notation = pd.DataFrame({'x': (list(np.arange(1, 100, .25)))})
func_list = [first_eq, second_eq, third_eq, fourth_eq, fifth_eq, sixth_eq, seventh_eq, eighth_eq]
label_list = ['O($1$)', 'O($log n$)', 'O($n$)', 'O($n log n$)', 'O($n^2$)', 'O($2^n$)', 'O($n!$)', 'O($n^n$)']
for n, each_func in enumerate(func_list):
    big_o_notation[f'y{n}'] = big_o_notation.x.apply(each_func)
    
    
for i,each_series in enumerate(big_o_notation):
    if each_series != 'x':
        plt.plot(big_o_notation.x, big_o_notation[each_series], label=label_list[i-1])

plt.ylim(1.0,100.0)
plt.xlim(1.0,20.0)
plt.suptitle('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.text(8, 100, 'Charles Atchison', fontdict={'fontsize': 20, 'fontstyle': 'italic'})
plt.legend()
plt.show()


```
